### Grab Tweets from accounts

```python
all_accounts_tweets = []
```

#### Influencer accounts

```python
account_count = 1
for handle in influencer_accounts[14:]:
    print(handle + ' - account number ' + str(account_count))
    account_count += 1
    try:
        tweetCriteria = got3.manager.TweetCriteria().setUsername(handle.replace('@',''))#.setMaxTweets(10)
        tweets = got3.manager.TweetManager.getTweets(tweetCriteria)

        try:
            tweet_counter = 0
            for tweet in tweets:
                account_tweet_dict = {'tweet_text' : tweet.text,
                             'user' : handle,
                             'date' : tweet.date,
                             'retweets' : tweet.retweets,
                             'influencer' : 1,
                             'exchange' : 0,
                             'project' : 0,
                             'news' : 0}

                all_accounts_tweets.append(account_tweet_dict)
                tweet_counter += 1
                
                if tweet_counter % 10 == 0:
                    print(f"{tweet_counter} tweets grabbed from account: {handle}")

        except:
            print('An exception has occurred type 1')

    except:
        print('An exception has occurred type 2')
```

```python
all_tweets_df = pd.DataFrame(all_accounts_tweets)
```
